```python
import pandas as pd
```

Creating DataFrames and fillig them with the two CSV files. 

```python
df1 = pd.read_csv("PepsiCo_FMCG_Dim.csv")
```

```python
df2 = pd.read_csv("PepsiCo_FMCG_Fact.csv")
```

Finding out the basic info of the two data frames.

```python
df1.info()
```

```python
df2.info()
```

Checking for duplicates

```python
duplicated_rows = df1[df1.duplicated]
```

```python
print(duplicated_rows)
```

```python
duplicated_rows = df2[df2.duplicated()]
```

```python
print(duplicated_rows)
```

Checking for nulls in each column and summing them 

```python
df1.isnull().sum()
```

```python
df2.isnull().sum()
```

Dropping the rows with nulls in ProductID, UnitsSold, SellPrice

```python
df2 = df2.dropna(subset= ['ProductID'])
df2 = df2.dropna(subset= ['UnitsSold']) 
df2 = df2.dropna(subset= ['SellPrice'])
```

Checking for nulls in each column and summing them after dropping the nulls from few columns

```python
df2.isnull().sum()
```

Replacing the nulls in the Region column with "Other Region"

```python
df2.fillna({'Region' : 'Other Region'},inplace=True)
```

Checking for nulls in each column and summing them after dropping the nulls from few columns, again

```python
df2.isnull().sum()
```

We've successfully eliminated all the nulls

Exporting the latest dataframes to CSV

```python
df1.to_csv("cleaned_PepsiCo_FMCG_Dim.csv", index = False)
df2.to_csv("cleaned_PepsiCo_FMCG_Fact.csv", index = False)
```
